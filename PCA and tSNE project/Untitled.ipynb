```python
thedata = pd.DataFrame([[120,302,403,204],[13,23,32,44],[12,21,30,44],[17,28,39,42]])
thedata = pd.DataFrame([[10,100,10,100],[200,20,2000,20],[30,30,300,300],[40,40,40,40]])
#thedata.plot()
scaler=StandardScaler()
data_scaled=pd.DataFrame(scaler.fit_transform(thedata), columns=thedata.columns)

data_scaled
#data_scaled.plot()


n=data_scaled.shape[1]
pca = PCA(n_components=n, random_state=1) #Apply the PCA algorithm with random state = 1

data_pca1 = pd.DataFrame(data = data_scaled ) #Fit and transform the pca function on scaled data
pca.fit_transform(data_pca1)
exp_var = pca.explained_variance_
#pd.DataFrame(exp_var).cumsum().plot()



# visualize the explained variance by individual components
plt.figure(figsize = (10,10))
plt.plot(range(1,5), exp_var.cumsum(), marker = 'o', linestyle = '--')
plt.title("Explained Variances by Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
```
