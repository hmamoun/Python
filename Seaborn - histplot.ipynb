```python
penguins.head()
penguins.shape
penguins.dropna(inplace=True)


penguins.shape
```

```python
sns.set_style('darkgrid')


sns.histplot(penguins.bill_depth_mm)
penguins.sex.unique()
```

```python
sns.histplot(data=penguins, x="body_mass_g" , 
             #hue="sex",
            stat='probability',
             cumulative = True
            )
```

```python
#penguins = penguins[((penguins.body_mass_g >= 3800) & (penguins.body_mass_g <=4000 )) ]

#print(penguins , penguins.info())

pf = penguins[(penguins.sex=='Female')]
pm = penguins[(penguins.sex=='Male')]
sns.histplot(data=pf, x="body_mass_g" , hue="sex" , color='indigo',
             kde = True,
            # bins=[3000 , 4500 , 4500,6000]             
            )

```

```python
sns.histplot(data=pm, x="body_mass_g" , hue="sex" ,  color='g',
            kde = True,
            # bins=[3000 , 4500 , 4500,6000]            
            )
```

```python
sns.histplot(data=penguins, x="body_mass_g" , 
             #hue="species",
             hue='sex',
            stat='count',
             #element='poly',
             multiple='fill',
             shrink= 0.8
             
            )
```

```python
sns.histplot(data=penguins , x = 'bill_length_mm' ,y='species',
            cbar = True,
            hue='sex',  
             multiple='dodge',
             palette = 'bone'
             
            )
```
