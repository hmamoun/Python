```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#def draw_plot():
    # Read data from file
df =  pd.read_csv("epa-sea-level.csv")
    # Create scatter plot
df.head()

fig, axes = plt.subplots( )
fig = plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'] ,  c="g", alpha=0.5)
axes.set_ylabel("CSIRO Adjusted Sea Level")
    
    # Create first line of best fit
fst = linregress(df.Year ,df['CSIRO Adjusted Sea Level'])  

xp = pd.Series([i for i in range (1880,2050)])
yp = fst.slope*xp + fst.intercept
plt.plot(xp , yp )
    # Create second line of best fit
mask = (df["Year"] >=2000)
filteredDF = df[mask]

fst2 = linregress(filteredDF.Year ,filteredDF['CSIRO Adjusted Sea Level'])    
xp1 = pd.Series([i for i in range (2000,2050)])
yp1 = fst2.slope*xp1 + fst2.intercept
plt.plot(xp1 , yp1 )

    # Add labels and title
axes.set_xlabel('Year')
axes.set_ylabel('Sea Level (inches)')
axes.set_title('Rise in the Sea Level')
    
    # Save plot and return data for testing (DO NOT MODIFY)
    #plt.savefig('sea_level_plot.png')
    #return plt.gca()
```
