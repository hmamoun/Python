```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(x = cars.displacement)
```

```python
sns.violinplot(x = cars.cylinders , y=cars.displacement , hue=cars.origin)
```

```python
sns.violinplot(data=cars[cars.origin.isin(['japan' , 'europe'])] ,x ='cylinders' , y='displacement' , hue='origin'
              
              , split = True 
              #, inner = 'quartiles'
              , scale = 'count', scale_hue=False
              )
plt.legend(loc=2)
```

```python
cars[cars.origin.isin(['japan' , 'europe'])].groupby('cylinders').origin.value_counts()
```
