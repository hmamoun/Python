### Prediction model regression

I will use a regression model to predict the future engaments:
- I used the data of the different post types (video, picture, status or link), because they don't behave the same way
- I removed the outliers between 95% and 100% as it was obvious we unexplained spikes
- I only used the data starting 2016, as the historical data before that is out of context due to the dramatic dropdown in the users interactions beginning of 2016
- the prediction for each data type is stated is a linear formula using the (t) variable which is the number of days after 2016-01-01 


```python
df
```

```python
from sklearn.linear_model import LinearRegression
df['date'] = df['status_published_date']

def predict(postType,reaction='num_likes'):
    datemask = (df.year >= 2016) ## just to make sure we're consuming the right period of the data
    mask = (df.status_type==postType)
    vdf = df[datemask]
    vdf = vdf[mask]
    vdf = vdf[['status_type','date','all_interactions' , 'year' , reaction]].groupby('date').mean()
    #vdf = vdf[vdf["all_interactions"] < 3000]
    vdf['avg_interactions'] = vdf[reaction]
    #vdf.drop(labels='avg_interactions' , inplace = True, axis=1)
    P = np.percentile(vdf.avg_interactions, [0, 95])
    vdf = vdf[(vdf.avg_interactions > P[0]) & (vdf.avg_interactions < P[1])]
    vdf.sort_index(inplace=True)
    vdf.reset_index(level=0, inplace=True)
   
    model = LinearRegression().fit(np.array(vdf.index).reshape(-1, 1),vdf.avg_interactions) 
    coefficients = [model.coef_[0],model.intercept_] 

    print("The linear trend is given for " +reaction + " for " + postType + " by F(t) = " +str(coefficients[0])+"*t + (" +str (coefficients[1])+")") 
    print("(t) is number of days of the post starting " , vdf['date'].min().date())
    linear_vdf = model.predict(np.array(vdf.index).reshape(-1, 1)) 

    plt.plot(vdf.index, vdf.avg_interactions, color = 'navy', label = 'Original Trend')
    plt.xlabel('No of days')
    plt.ylabel(postType + ' Interactions')
    plt.title('Average ' + postType + ' Ineractions')
    plt.plot(vdf.index, linear_vdf, color = 'orange', label = 'Linear Trend')
    plt.legend(loc = 'upper right')
    plt.show()
```

```python
predict('video','num_loves')
```
