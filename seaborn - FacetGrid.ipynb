```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
penguins = sns.load_dataset('penguins')
sns.get_dataset_names()
```

```python
penguins.head()
```

```python
penguins.head()
sns.set_style('white')
g = sns.FacetGrid(data = penguins , col='island' , row='sex'             )
g.map(sns.histplot , "flipper_length_mm")
```

```python
g = sns.FacetGrid(data = penguins , col='island', row='sex'             )
g.map_dataframe(sns.histplot , y='flipper_length_mm')
```

```python
g = sns.FacetGrid(data = penguins , col='island', row='species'  
                 ,sharey=False #Axis Y will not be shared
                 
                  ,ylim=(20,70) #set the limits of the axis
                 )
g.map_dataframe(sns.scatterplot ,  x = 'bill_depth_mm',y='bill_length_mm' )
g.set_axis_labels("Bill Depth","Bill length")
g.set_titles(col_template='{col_name} Island')
```

```python
g = sns.FacetGrid(data = penguins , col='island', row='species'  
                 ,sharey=False #Axis Y will not be shared
                  ,hue='sex'
                  ,ylim=(20,70) #set the limits of the axis
                 )
g.map_dataframe(sns.scatterplot ,  x = 'bill_depth_mm',y='bill_length_mm')

g.add_legend()
```

# Using custom function with FacitGrid

```python
def add_mean_line(data , var=None , **kws): #Must have an argument called "data"
    #if not varibale provided skip adding mean line
    if not var: return
    #calculate mean for each group
    m=np.mean(data[var])
    #get current axis
    ax=plt.gca()
    #add line at group mean
    ax.axvline(m, color='maroon',lw=3,ls='--')
    
    #annotate group mean
    x_pos = 0.65
    if m>5000: x_pos=0.2
    ax.text(x_pos,0.7,f'mean={m:.0f}', transform=ax.transAxes,
           color='maroon' ,fontweight='bold',fontsize=12)
```

```python
quantity = 'body_mass_g'
g = sns.FacetGrid(penguins,row='sex' , col = 'species')
g.map_dataframe(sns.kdeplot , x=quantity , lw=3 , shade=True)
g.map_dataframe(add_mean_line,var=quantity)
g.set_titles(row_template='{row_name} is awesome' , col_template = 'Same as {col_name}')
```
